```python
import requests
import pandas as pd
from io import StringIO
from tqdm import tqdm
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
```

```python
base = "https://stats.ncaa.org"
allteams = pd.read_html("https://stats.ncaa.org/selection_rankings/nitty_gritties/37268",extract_links='body')[0]
```

```python
allRosters
```

```python
allRosters = pd.DataFrame()
for team, teamid in tqdm(allteams.Team):
    roster = pd.read_html(base + teamid + '/roster')[0]
    roster['team']   = team
    roster['teamID'] = teamid.split('/')[2]

    allRosters = pd.concat([allRosters, roster])
```

```python
# Save games data
allRosters.to_csv('rosters_2023_24.csv',index=False)
```
